## import hashlib组函数

- hash基础
- hash的参数必需为bytes(字节或字节串的类型都是bytes)(空字节串b""也是)，非要用字符串的话需要后面加上编码格式例：hashlib.md5("hello".encode("utf-8"))才行
- b"hello" 前面的b标明这是二进制类型，其储存的信息是"hello"字符串，b"hello"的类型是bytes，而"hello"类型是str(字符串)，两者不同

## hashlib.---
- hashlib.---(eg:md5，即hashlib.md5，md5只是计算器的一种运算规则，也有其它规则能使原信息储存为不同的大小内容)
- hashlib.md5中储存的信息并不是原文本信息，例如一个百万字的文本文档经过hashlib.md5()函数处理后hashlib.md5只保存32个字节，因此经由hashlib.md5储存的信息是不可逆的，即不可通过hashlib储存的信息还原出原文本信息，另外，除了md5这种规则返回的是32个字节，还有其它的运算规则，例如hashlib.sha256()中的.sha256规则下返回的字节数就比32个字节多
- .hexdigest()的作用，作用于例：hashlib.md5.hexdigest()，将储存的md5个形式的数据经过处理返回特定的字符串，相比于直接保存文本，图片，视频等内容占用大量内存，以及不能通用保存图片，ppt等各种格式，用hashlib.md5.hexdigest()能应对前面这些情况，用后续.hexdigest()转换的原因是相比于二进制的字节串数据，字符串更便于后续写语句
- ps：例：hashlib.sha256(b"hello").hexdigest()经.hexdigest()处理后返回的是'5d41402abc4b2a76b9719d911017c592'
-
##  注意：
- hashlib.md5 是创建 md5 对象的函数
- hashlib.md5() 会调用这个函数，创建一个新的 md5 对象
- md5 = hashlib.md5()表示用变量 md5 保存这个对象的引用。之后 md5.update(chunk) 会不断更新同一个对象内部的 hash 状态，如果不保存对象，而是反复 hashlib.md5().update(chunk)，每次都会创建新对象，状态无法连续累积，因此不能得到整个文件的 hash

## 不用读取的二进制直接判断相等而用hashlib.---.hexdigest()判断相等的原因
- 实际上，内容一样的文件(包括txt,ppt,mp4等)的二进制内容是相等的，可以用来判断相等，只是这些二进制数据记录着文件的全部信息，而用全部信息直接比较的话会占用较大的内容，例如比较两个几G大小的视频直接用二进制就会占用较大的内存甚至出现错误(即占用内存多于运行电脑的实际内存，而用hashlib函数处理后如md5规则返回的就是32个字节，用来比较效果就好一点
-
- 只不过直接用hashlib而不用.hexdigest()显示的话，==判断的是是否是同一个对象，因此两个hashlib.md5里放的内容相同的变量==的判断结果为False，而加上.hexdigest()显示后，==判断的是内容或值是否相同，所以两个hashlib.md5里放的内容相同的不同变量的==的判断结果为True

## 对象.update函数
- 每一个对象里都有当前的hash状态，当使用.update这种需要对同一个对象进行多次操作的函数时，需要先创建一个对对象的引用，而不是直接调用函数创建对象，否则，虽然.update不会报错，但不保存对象引用，就无法持续操作同一个对象，也无法读取最终结果。(因为一般用时是想对同一个对象进行多次操作，而直接调用函数创建对象多次操作时相当于每次都创建了一个新的对象，这导致变成了每次对不同对象进行了一次操作，较容易与理解的有出错，所以python选择直接进行报错，增强了规范性)
- .update函数，相当于记录书签，每次调用时返回所传的内容并标记所传对象，在算传内容的末尾处定一个书签，当作用于同一个对象时，每次对象.update()会从上一次所传入的对象的内容的光标处往下继续进行所传的内容返回给.前面的函数，而不是重新从开始处进行
- ps：即像读书时加一个书签，下次读书时从上次的书签处接着往下读，而不是又重新开始读
-
- ps：.read也是像读书时加一个书签，下次读书时从上次的书签处接着往下读，而不是又重新开始读

## hashlib中的有关hash算法简单呈现
- 实际上，在用update()处理添加的循环过程中，md5保存整个 Hash 计算时的中间状态（Internal State），即根据md5对象所储存的传入的值经过halishlib.md5()返回的计算器处理后的状态
-
- 关于hashlib.md5函数，它进行处理时是有一个初始状态，以及每次添加新的字节(注：添加空内容以及添加空字节时并不会导致hash值改变)会根据所加的字节的内容进行复杂的运算最后修改运算状态，因为空字节与空内容并不会导致hash值改变，所以对空内容和空字节处理时的结果(hashlib.md5()和hashlib.md5(b""))是相同的，则用.hexdigext()显示后用==判断的结果是True
