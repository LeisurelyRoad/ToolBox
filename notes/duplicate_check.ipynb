## 根据文件大小分组

- 列表用于处理储存信息，尤其是储存统计信息
- 字典用于统计需要分层信息，例如这一层那一层，每一层对应的一个key
- 以分层为主主用字典
- 以储存为主主用列表
- 在没有必要分组时就用列表
- 另外，在判断遍历后是否存在为达成遍历目的时，往往可以通过列表或字典的建立位置，利用遍历后均不达成时列表或字典为空的情况来判断

## with open(file_path,"rb")as f: 中的rb

- with open(file_path,"rb")as f:中的rb：r->(read)读取,b->(binary)二进制，rb即以二进制方式读取文件，因为不进行从字节到文字的转换，所以后面不用写encoding
- 当同时处理不同类型的文件如ppt,txt,mp4,jpg等用rb以二进制方式读取文件可以避免用encoding=""时只有部分文件支持的情况

## import hashlib组函数

- hash基础
- hash的参数必需为bytes(字节或字节串的类型都是bytes)(空字节串b""也是)，非要用字符串的话需要后面加上编码格式例：hashlib.md5("hello".encode("utf-8"))才行
- b"hello" 前面的b标明这是二进制类型，其储存的信息是"hello"字符串，b"hello"的类型是bytes，而"hello"类型是str(字符串)，两者不同
-
## hashlib.---
- hashlib.---(eg:md5，即hashlib.md5，md5只是计算器储存内容的一种格式，也有其它格式能使原信息储存为不同的大小内容)
- hashlib.md5中储存的信息并不是原文本信息，例如一个百万字的文本文档经过hashlib.md5()函数处理后hashlib.md5只保存32个字节，因此经由hashlib.md5储存的信息是不可逆的，即不可通过hashlib储存的信息还原出原文本信息，另外，除了md5这种储存形式，还有其它的储存形式，例如hashlib.sha256()中的.sha256储存的就比32个字节多
- .hexdigest()的作用，作用于例：hashlib.md5.hexdigest()，将储存的md5个形式的数据经过处理返回特定的字符串，相比于直接保存文本，图片，视频等内容占用大量内存，以及不能通用保存图片，ppt等各种格式，用hashlib.md5.hexdigest()能应对前面这些情况，用后续.hexdigest()转换的原因是相比于二进制的字节串数据，字符串更便于后续写语句
- ps：例：hashlib.sha256(b"hello").hexdigest()经.hexdigest()处理后返回的是'5d41402abc4b2a76b9719d911017c592'
-
## 不用读取的二进制直接判断相等而用hashlib.---.hexdigest()判断相等的原因
- 实际上，内容一样的文件(包括txt,ppt,mp4等)的二进制内容是相等的，可以用来判断相等
