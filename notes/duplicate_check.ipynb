## 根据文件大小分组

- 列表用于处理储存信息，尤其是储存统计信息
- 字典用于统计需要分层信息，例如这一层那一层，每一层对应的一个key
- 以分层为主主用字典
- 以储存为主主用列表
- 在没有必要分组时就用列表
- 另外，在判断遍历后是否存在为达成遍历目的时，往往可以通过列表或字典的建立位置，利用遍历后均不达成时列表或字典为空的情况来判断
