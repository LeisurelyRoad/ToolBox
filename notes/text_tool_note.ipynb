# 文本处理笔记

## 1. 查看txt内容

- 整体用.read()返回字符串，并且只调用一次print省去了每次用print时自带的换行而额外处理\n的操作；
- 分行处理用.readlines()，按每行返回字符串为列表元素，整体处理大文件时(>=1GB)用readlines()处理减少内存占用，提升运行速度
-
- try except：先尝试出错再处理，用于执行一个可能失败的操作；用户层还可以提示用户出错部分原因；
- if os.path.exists()用于逻辑判断

## 2.统计字数和行数

- 处理和显示分离(处理函数尽量返回处理数据，而不是直接打印；ps：本身功能就是打印的函数除外)；函数返回数据时，返回的其实是一个元组(a,b,...)(a,b等为元组元素)
-
- 在使用处理和呈现的两种函数实现同一个功能时，可在呈现函数处使用try (调用处理函数), except来处理处理函数可能出现的错误
-
- 用.splitlines()可以将文本内容按照每行为一个列表元素返回列表，但是它对特殊的边界情况有相应的规则："text = '' text.splitlines()返回的是[]，len()计算的长度为0"；"text = "hello/nworld/n" text.splitlines()返回的是['hello','world']，len()计算的长度为2"；"text = "hello/nworld" text.splitlines()返回的也是['hello','world']，len()计算的长度为2"
- 在设计程序时，应根据用户认为什么是"一行"，进而确定程序中相应的行数定义

-len()在计算字符串时返回的是字符的个数，可以看作是将字符串的字符看为字符串的元素,(ps:a = "Hello" a[1]：表示字符e)，只不过字符串不可修改

- char_count = len(text_content.replace("\n","").replace("\r",""))，去掉text_content的换行，""与" "不同，一个是无元素，一个是空格元素，前者不被len统计，后者被len统计
- 例如：a = "" print(len(a))打印的结果是0

## 统计关键字出现次数

 - .split().前面的变量字符串中遇见("")内容后将其删去，并将两侧的内容分为元素，例如text = "abdkdefkghi" print(text.split("k"))的结果是['abc','def','ghi']
