## with open() 中第一个变量的问题

- 分为：
- 绝对路径：直接输入完整的路径字符串(包括变量)
- 相对路径：../表示当前运行程序的目录的上一级目录，./表示当前运行程序的目录的同级目录
- 文件名： 相对路径一种简写，相当于./文件名，即取本级目录
- 注意：这里的相对路径是基于当前的工作目录的，也就是即使例如backup.txt是在file_tools.py文件定义的，但运行时是在main.py运行的，因此是以main.py所在的ToolBox目录为基准对标的
- print(os.getcwd())：打印当前的工作目录，例如本项目的工作目录是......./阶段三：软件工程基础/ToolBox
-
- os.path.abspath("")将括号里的相对路径字符串改为绝对路径字符串
- 用于下一级路径的相对路径表示,找到工作目录的本级对应的所找的文件所在的文件夹加/文件名和等类似方法 例如若将backup.txt放到tools的text文件夹里,则用"tools/text/backup.txt"来表示相对路径进行寻找

- 字典中的key不同时才会新增，当是相同的key时，只是将字典的key指向的value值换为了新的value值
-
- 列表里可以放字典作为列表的一个元素，字典中的key的格式需为字符或字符串的形式
-
- 打印整个字典时，字典中引用字符或字符串的引号打印在界面上显示的是单引号，只打印字典的key对应的value值时，会去掉表示引用的双引号打印出来

## 不需要保存的数据不保存

- 数据越多,维护成本越高,出现不一致的风险越高(如今最主要的原因之一)只保存真正需要长期保存的数据，可以推导出来的与后续计算关联度不高以及实现难度简单的数据根据情况考虑是否重复保存。
- 内存占用变高，运行速度下降(一般而言，1000个文本语句的内存占用可以忽略)

## 工程维护

- 降低复杂度,降低维护成本->减少特殊情况,减少重复代码,让程序更通用,不需要保存的数据不保存
