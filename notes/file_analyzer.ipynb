## 统计后缀文件数量

- 用suffix来替代file_data["suffix"]，是因为这样当遇见没有后缀的情况下执行if suffix == "":suffix = "无后缀"对suffix_count字典进行了修改，因此，在后面不常用修改原始量或修改原始量的步骤简单的情况下都要用新建局部变量来代替修改原始量的语句，这是一种安全思维

- suffix_count = suffix_count.get(suffix, 0)+1 是if suffix not in suffix_count: ， suffix_count[suffix] = 0，suffix_count[suffix] += 1语句的简洁版，这个语句是作用于字典suffix_count，括号里逗号前面填的是key,后面填的是调用出错时的返回值，返回值可以是字符，字符串，数字，列表，字典，等各种形式，这里即表明当在suffix_count字典查找key->suffix时，发现没有suffix这个key，调用出错，返回括号里的数字0，即suffix_count.get(suffix, 0)的初始值为0，0+1=1,这时suffix_count.get(suffix, 0)里有suffix，且此key对应的值是1，再执行就是1+1=2，即类似于之前的+=的逻辑了

- [:n]是列表切片，即切到从第n个元素，将前面包括第n个元素作为新列表返回，例如a=[1,2,3,4,5] print(a[:3]) 则打印[1,2,3]

- enumerate()
- for i , x in enumerate(名字(这个参数只要是可迭代对象就行),start=i(这里的i表示从第i个元素开始)如果没有start=i则默认是start=0
- i和x都是变量名都可以修改，这里暂时将i改为I来说明，I是start=1时，I第一次等于1,之后每次I=I+1,如果是start=i的情况，那第一次I=i，之后每次I=I+1
- x则是类似于for x in a 的顺序每次对应的列表元素,i 和 x 是由同一个元组解包得到的两个变量，
修改其中一个变量，不会影响另一个变量，这里的i和x类似于解包，互相之间并不影响
- 例如：
- a = ["a", "b", "c"]
-
- for i, x in enumerate(a):
-    print(i, x)
- 结果是
- 0 a
- 1 b
- 2 c
